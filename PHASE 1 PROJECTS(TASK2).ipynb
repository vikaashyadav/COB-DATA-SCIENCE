```python
from scipy import stats

# Upload data from a CSV file 
df = pd.read_csv('cleaned_data.csv')

# Define a function to identify outliers based on z-score
def find_outliers(df, z_threshold=2):
    outlier_info = {}
    
    for column in df.columns:
        if df[column].dtype in [np.float64, np.int64]:  # Only numeric columns
            z_scores = np.abs(stats.zscore(df[column]))
            outliers = df[z_scores >= z_threshold][column].tolist()
            if outliers:
                outlier_info[column] = outliers
    
    return outlier_info

# Find columns with outlier values
z_threshold = 2  
outlier_info = find_outliers(df, z_threshold)

# Display columns with outliers and their outlier values
for column, outliers in outlier_info.items():
    print(f"Column '{column}' has outliers:")
    print(outliers)
```

```python
# Define a function to identify outliers based on z-score and remove them
def remove_outliers(df, z_threshold=2):
    cleaned_df = df.copy()
    
    for column in df.columns:
        if df[column].dtype in [np.float64, np.int64]:  # Only numeric columns
            z_scores = np.abs(stats.zscore(df[column]))
            outliers_mask = z_scores >= z_threshold
            cleaned_df = cleaned_df[~outliers_mask]
    
    return cleaned_df

# Outlier values remove them to get the cleaned dataset
z_threshold = 2  
cleaned_df = remove_outliers(df, z_threshold)

# Display the cleaned dataset
print("Cleaned DataFrame (outliers removed):")
print(cleaned_df)
```
